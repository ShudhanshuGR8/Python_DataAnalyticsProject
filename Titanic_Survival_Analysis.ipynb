# **TITANIC SURVIVOR ANALYSIS**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(pd.read_csv('/content/train (1).csv'))
df.head()
```

```python
df.shape
```

This shape function is the analysis that there are 891 rows and 12 columns in the DataFrame.

```python
df.isnull().sum()
```

The aforementioned analysis is done in order to handle NULL values that might be a part of our DataFrame.

```python
drop_col = df.isnull().sum()[df.isnull().sum()>(35/100 * df.shape[0])]
drop_col
```

In the Above code snippet we are actually separating the columns from our DataFrame which have more than 35 % of their values as NULL.

```python
drop_col.index
```

```python
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

Embarked Column here contains String values so mean was not able to be calculated hence we are still seeing that it has two null values.

```python
df['Embarked'].describe()
```

```python
df['Embarked'].fillna("S", inplace=True)
df.isnull().sum()
```

Finally All the NULL values have been dealt with.

```python
df.corr()
```

```python
df['FamilyMan'] = df['SibSp'] + df['Parch']
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df.corr()
```

Above, we created a new Column 'FamilyMan' by combining the two Columns 'SibSp' and 'Parch'.

Quick Conclusion: THe Survival Rate is not that greatly influenced by the FamilyMan i.e. the Survival Rate is irrespective of whether the Family is big or small.

```python
df['SinglePerson'] = [0 if df['FamilyMan'][i]>0 else 1 for i in df.index]
df.head()
```

Here 0 indicates False and 1 indicates True.

```python
df.groupby(['SinglePerson'])['Survived'].mean()
```

Well, the Above Data Analysis concludes that a Single Person or one is travelling Alone has lesser or fewer chance of Survival as a FamilyMan who was able to affor the ticket for his/her family shows that he/she would have been Rich and we all know how the Rich gets prioritised over all.

```python
df[['SinglePerson', 'Fare']].corr()
```

Quick Conclusion: For someone not a SinglePerson, the chance of the Fare being Higher is High.

```python
df['Sex'] = [0 if df['Sex'][i]=='male' else 1 for i in df.index] 
#Note: 0 is for Male and 1 is for FeMale
df.groupby(['Sex'])['Survived'].mean()
```

It is clearly visible that Female Passengers have more chance of survival than the Male ones. It seems that Female Passengers were prioritised over the Male Passengers.

```python
df.groupby(['Embarked'])['Survived'].mean()
```
